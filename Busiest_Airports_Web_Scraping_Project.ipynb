```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_busiest_airports_in_India'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup.prettify())
```

```python
table = soup.find_all('table')[1]
```

```python
col_titles = table.find_all('th')
col_titles
```

```python
col_table_titles = ['Rank','Name','City','State/UT','IATA Code','Passengers in 2022-23','Passengers in 2021-22','% Change','Rank Change']
```

```python
df = pd.DataFrame(columns=col_table_titles)
```

```python
column_data = table.find_all('tr')
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
df.to_csv(r'airports.csv', index = False)
```
